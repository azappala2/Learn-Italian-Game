```python
# anthony zappala January 1st, 2024
# I want to learn Italian so I made this flashcard game to help me learn
# I also want to get better at python so I made this game to practice
# I used the tkinter module to make the GUI


from tkinter import *
from random import randint
import random as random
from datetime import date
from datetime import datetime
root = Tk()
root.title('Italian Language Flashcards')
root.geometry("700x500")

global todays_date
todays_date = date.today()
global date_formatted
date_formatted = todays_date.strftime("%B %d, %Y")

global current_time
current_time = datetime.now()
global time_formatted
time_formatted = current_time.strftime("%H:%M")

global status 
status = ""

#verb flashcard function
def verbs():
    #hide previous frame
    hide_all_frames()
    verb_frame.pack(fill="both", expand=1)
    global status 
    status = "verbs"
    #create a list of verbs
    verbs = [
    (("avere"), ("to have")),
    (("essere"), ("to be")),
    (("andare"), ("to go")),
    (("fare"), ("to do")),
    (("dire"), ("to say")),
    (("partire"), ("to leave")),
    (("stare"), ("to stay")),
    (("dare"), ("to give")),
    (("vedere"), ("to see")),
    (("sapere"), ("to know")),
    (("volere"), ("to want")),
    (("scrivere"), ("to write")),
    (("capire"), ("to understand"))
    ]
    
    #keep track of score
    global score 
    score = 0
 
    def nextWord():
        global random_word 
        global hinter
        global hint_count  
        
        #clear screen
        my_entry.delete(0, END)
        answer_label.config(text="")
        hint_label.config(text="")
        hinter = ""
        hint_count = 0

        #pick new random word
        random_word = randint(0, len(verbs)-1)
        italian_word.config(text=verbs[random_word][a])
        

    def checkAnswer():
        if my_entry.get().lower() == verbs[random_word][b]:
            answer_label.config(text=f"Correct! {verbs[random_word][a]} is {verbs[random_word][b]}")
            global score
            global hint_count
            if hint_count == 0:
                score += 1
                Score.config(text = f" Score: {score}")
        else:
            answer_label.config(text="Incorrect!")

    #Keep track of hints
    global hinter 
    hinter = ""
    global hint_count 
    hint_count = 0
    def hint():
        global hint_count
        global hinter
        if hint_count < len(verbs[random_word][b]):
            hinter += verbs[random_word][b][hint_count]
            if hint_count == 0:
                hint_label.config(text=f"The first letter is {verbs[random_word][b][0]}")
            else:
                hint_label.config(text=f"The first {hint_count+1} letters are {hinter}")
            hint_count += 1
        


    italian_word = Label(verb_frame, text="", fg = "white", font=("Helvetica", 48))
    italian_word.pack(pady=10)


    answer_label = Label(verb_frame, text="")
    answer_label.pack()

    my_entry = Entry(verb_frame)
    my_entry.pack(pady=10)

    hint_label = Label(verb_frame, text="")
    hint_label.pack()

    button_frame = Frame(verb_frame)
    button_frame.pack(pady=10)


    
    answer_button = Button(button_frame, text="Check Answer", command=checkAnswer)
    answer_button.grid(row=6, column=0, pady=10, padx=10)

    hint_button = Button(button_frame, text="Hint", command = hint)
    hint_button.grid(row=6, column=1, pady=10, padx=10 )

    next_button = Button(button_frame, text="Next Word", command=nextWord)
    next_button.grid(row=6, column=2, pady=10, padx=10)

    Score = Label(button_frame, text=f"Score: {score}")
    Score.grid(row=7, column=1, pady=10, padx=10)
    
    nextWord()


def start_Phrase():
    global score1
    global a, b
    score1 = 0
    #phrase flashcard function
    def phrases():
        global status 
        global score1
        global wrong_guess
        global a,b 
        global random_index
        global answer_list

        status = "phrases"
        wrong_guess = 0
        #hide previous frame
        hide_all_frames()
        phrases_frame.pack(fill="both", expand=1)    

        #create a list of phrases in english
        eng_phrases = [
            "How are you?", 
            "I am good",
            "What is your name?",
            "My name is ____",
            "Where are you from?",
            "I am from ____",
            "What time is it?",
            f"It is {time_formatted}",
            "What is the weather like?",
            "It is hot outside",
            "It is cold outside",
            "It is raining",
            "It is snowing",
            "It is sunny",
            "What is the date?",
            f"Today is {date_formatted}",
            "What is your favorite color?"
        ]

        #the same list of phrases but in italian
        ital_phrases = [
            "Come stai?",
            "Sto bene",
            "Come ti chiami?",
            "Mi chiamo ____",
            "Di dove sei?",
            "Sono di ____",
            "Che ora è?",
            f"Sono le {time_formatted}",
            "Che tempo fa?",
            "Fuori fa caldo",
            "Fuori fa freddo",
            "Sta piovendo",
            "Sta nevicando",
            "C'è il sole",
            "Che data è oggi?",
            f"Oggi è il {date_formatted}",
            "Qual è il tuo colore preferito?"

            ]
        
        #initialize answer list
        answer_list = []

        #intialize the phrase label
        english_phrase = Label(phrases_frame, text="", fg = "white", font=("Helvetica", 48))
        english_phrase.pack(pady=10)

        count = 0

        #check which option is selected
        if a == 0:  
            #show the italian phrase, answers should be in english
            #make a dictionary with italian phrases as keys and english phrases as values
            combined_phrases = dict(zip(ital_phrases, eng_phrases))
            #print the random phrase
            random_index = randint(0, len(ital_phrases)-1)
            answer = ital_phrases[random_index]
            english_phrase.config(text=ital_phrases[random_index])
            while count < 4:
                if count == 0:
                    #add the random english phrase to the list of answer choices
                    answer_list.append(eng_phrases[random_index])
                    eng_phrases.remove(eng_phrases[random_index])
                    count += 1
                else:
                    #choose a random phrase from the initial english lists
                    random_index = randint(0, len(eng_phrases)-1)
                    #add the random phrase to the list of answer choices
                    answer_list.append(eng_phrases[random_index])
                    eng_phrases.remove(eng_phrases[random_index])
                    count += 1

        elif a == 1:
            #show the english phrase, answers should be in italian
            #make a dictionary with english phrases as keys and italian phrases as values
            combined_phrases = dict(zip(eng_phrases, ital_phrases))
            #print the random phrase
            random_index = randint(0, len(eng_phrases)-1)
            answer = eng_phrases[random_index]
            english_phrase.config(text=eng_phrases[random_index])
            while count < 4:
                if count == 0:
                    #add the random italian phrase to the list of answer choices
                    answer_list.append(ital_phrases[random_index])
                    ital_phrases.remove(ital_phrases[random_index])
                    count += 1
                else:
                    #choose a random phrase from the initial italian list
                    random_index = randint(0, len(ital_phrases)-1)
                    #add the random phrase to the list of answer choices
                    answer_list.append(ital_phrases[random_index])
                    ital_phrases.remove(ital_phrases[random_index])
                    count += 1


        random.shuffle(answer_list)
    
            
        button_frame1 = Frame(phrases_frame)
        button_frame1.pack(pady=10)
        
        global phrase_radio 
        phrase_radio = StringVar()
        phrase_radio.set(answer_list[0])
        phrase_radio_button1 = Radiobutton(button_frame1, text=answer_list[0], value=answer_list[0], variable=phrase_radio).pack()
        phrase_radio_button2 = Radiobutton(button_frame1, text=answer_list[1], value=answer_list[1], variable=phrase_radio).pack()
        phrase_radio_button3 = Radiobutton(button_frame1, text=answer_list[2], value=answer_list[2], variable=phrase_radio).pack()
        phrase_radio_button4 = Radiobutton(button_frame1, text=answer_list[3], value=answer_list[3], variable=phrase_radio).pack()

        phrase_answer_label = Label(button_frame1, text="")
        phrase_answer_label.pack(pady=10)
        #create phrase answer check
        def phraseCheck():
            global wrong_guess
            if phrase_radio.get() == combined_phrases[answer]:
                phrase_answer_label.config(text="Correct!")
                global score1
                if wrong_guess == 0:
                    score1 += 1
                    Score1.config(text = f" Score: {score1}")
            else:
                phrase_answer_label.config(text="Incorrect!")
                
                wrong_guess += 1


        phrase_check_button = Button(button_frame1, text="Check Answer", command=phraseCheck).pack(pady=10)
        
        
        next_phrase_button = Button(button_frame1, text="Next Phrase", command=phrases).pack(pady=10)

        Score1 = Label(button_frame1, text=f"Score: {score1}")
        Score1.pack(pady=10)
    phrases()

    
    
    

def hide_all_frames():
    #get rid of children from previous frames
    for widget in verb_frame.winfo_children():
        widget.destroy()
    for widget in phrases_frame.winfo_children():
        widget.destroy()
    for widget in opening_frame.winfo_children():
        widget.destroy()
    #hide frames
    opening_frame.pack_forget()
    verb_frame.pack_forget()
    phrases_frame.pack_forget()

def italianWord():
    global a, b
    a = 0
    b = 1
    global status
    if status == "verbs":
        verbs()
    elif status == "phrases":
        start_Phrase()
def englishWord():
    global a, b
    a = 1
    b = 0
    global status
    if status == "verbs":
        verbs()
    elif status == "phrases":
        start_Phrase()
        
def openframe():
    #hide previous frame
    welcome = Label(opening_frame, text="Welcome to the Italian Language Flashcard Game!", font=("Helvetica", 18))
    welcome.pack(pady=50)
    instructions = Label(opening_frame, text="Choose a game from the menu above to begin!", font=("Helvetica", 14))
    instructions.pack(pady=10)
    notice = Label(opening_frame, text = "Note: You may have to click around to access the answer box the first time", font=("Helvetica", 20))
    notice.pack(pady=30)
    opening_frame.pack(fill="both", expand=1)

#create menu
main_menu = Menu(root)
root.config(menu=main_menu)

#create menu items
verbs_menu = Menu(main_menu)
main_menu.add_cascade(label="Game Choice", menu=verbs_menu)
verbs_menu.add_command(label="Verbs", command = verbs)
verbs_menu.add_command(label="Phrases", command = start_Phrase)

#create second dropdown
options_menu = Menu(main_menu)
main_menu.add_cascade(label="Options", menu=options_menu)
options_menu.add_command(label="Show Italian Words", command = italianWord)
options_menu.add_command(label="Show English Words", command = englishWord)
#create frames
opening_frame = Frame(root, width=700, height=500)
verb_frame = Frame(root, width=700, height=500)
phrases_frame = Frame(root, width=700, height=500)



openframe()
italianWord()


root.mainloop()
```
